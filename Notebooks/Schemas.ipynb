```sql
-- CREATE SCHEMA WITHOUT EXTERNAL LOCATION IN DEV CATALOG
CREATE SCHEMA IF NOT EXISTS dev.bronze
COMMENT 'This is schema in DEV without external location';
```

```sql
-- CREATE SCHEMA WITHOUT EXTERNAL LOCATION IN DEV_EXT CATALOG
CREATE SCHEMA IF NOT EXISTS dev_ext.bronze
COMMENT 'This is schema in DEV EXT catalog without external location';
```

```sql
-- CREATE EXTERNAL LOCATION FOR SCHEMA
CREATE EXTERNAL LOCATION `ext_schema`
URL 'abfss://data@learndatabrickssg.dfs.core.windows.net/adb/schema'
WITH (STORAGE CREDENTIAL `uc_catalog_storage`);
```

```sql
-- CREATE SCHEMA WITH EXTERNAL LOCATION IN DEV_EXT CATALOG
CREATE SCHEMA dev_ext.bronze_ext
MANAGED LOCATION 'abfss://data@learndatabrickssg.dfs.core.windows.net/adb/schema/bronze_ext'
COMMENT 'This is schema in DEV_EXT catalog with external location';
```

```sql
-- CREATE TABLE RAW_SALES IN ALL THREE SCHEMA

CREATE TABLE DEV_EXT.BRONZE_EXT.RAW_SALE(
  id STRING,
  name STRING,
  PRICE DECIMAL(10,2),
  QUANTITY INT
);

-- INSERT DATA
INSERT INTO DEV_EXT.BRONZE_EXT.RAW_SALE VALUES ('1','Product1',100.00,100);
```
